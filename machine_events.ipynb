```python
# time column

import csv
d=dict()
job_failed=0
job_passed=0
with open('Machine_Events.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=',')
    for row in csv_reader:
        p=row[0]
        if p not in d:d[p]=1
        else:d[p]+=1
    for key,val in list(d.items()):
        if key=='0':job_failed=val
        else:
            job_passed+=1
    print("No of times job passed successful",job_passed)
    print("No of times job failed",job_failed)
```

The column 1 of the machine_events file contains time 
This column is mandatory field.
The format of field is integer.

```python
# CPU'S column
blank=0
d=dict()
with open('Machine_Events.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=',')
    for row in csv_reader:
        p=row[4]
        if p=='':
            blank+=1
            continue
# 32 colums are with blank cpu's data            
        if p not in d:d[p]=1
        else:d[p]+=1
    for key,val in list(d.items()):
        print(key,"   ",val)
    print("blank","   ",blank)
```
