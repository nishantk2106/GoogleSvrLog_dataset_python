## finding the date from the time column for the machine_events:

```python
from datetime import datetime
date=dict()
with gzip.open('/ichec/work/mucom001c/GoogleSvrLog/clusterdata-2011-1/machine_events/part-00000-of-00001.csv.gz','r') as f: 
    for line in f.readlines():
        line=line.decode()
        line=line.split(',')
        time=int(line[0])/1000
        date_time = datetime.fromtimestamp(time)
        d1=date_time.strftime("%d")
        if d1 not in date: date[d1]=1
        else:
            date[d1]+=1
print(date)
lists = sorted(date.items()) # sorted by key, return a list of tuples
x, y = zip(*lists) # unpack a list of pairs into two tuples
plt.figure(figsize=(20,10))
plt.bar(x, y,color='cyan')
plt.xlabel('Time as Date')
plt.ylabel('count')
plt.title('Plot for job on each date of the month of the cluster data ')
plt.show()       
        
```

## Number of event_type with different cpu capacity

```python
from matplotlib import pyplot as plt
ls=list()
with gzip.open('/ichec/work/mucom001c/GoogleSvrLog/clusterdata-2011-1/machine_events/part-00000-of-00001.csv.gz','r') as f: 
    for line in f.readlines():
        line=line.decode()
        line=line.split(',')
        evn_id=line[2]
        cpu_cap=line[4]
        ls.append((evn_id,cpu_cap))
        
x, y = zip(*ls) # unpack a list of pairs into two tuples
plt.scatter(x, y,color='red')
plt.xlabel('Event_type')
plt.ylabel('CPU_Capacity')
plt.title('Plot for Event_type Vs CPU_capacity ')
plt.show()
```

## Plotting graph to obtain the machine events associated with the time of date  :

```python
from datetime import datetime
ls=list()
n=0
with gzip.open('/ichec/work/mucom001c/GoogleSvrLog/clusterdata-2011-1/machine_events/part-00000-of-00001.csv.gz','r') as f: 
    for line in f.readlines():
        line=line.decode()
        line=line.split(',')
        time=int(line[0])/1000
        date_time = datetime.fromtimestamp(time)
        d1=date_time.strftime("%d")
        mac_id=line[1]
        if mac_id not in ls:
            ls.append((d1,n))
            n+=1
ls.sort()
x, y = zip(*ls) # unpack a list of pairs into two tuples
plt.figure(figsize=(15,8))
plt.scatter(x, y,color='cyan')
plt.xlabel('Time as Date')
plt.ylabel('Machine_id as number')
plt.title('Plot for Machine_id  Vs Time ')
plt.show()    
```
