```python
# time column

import csv
d=dict()
job_failed=0
job_passed=0
with open('Task_Constraints.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=',')
    for row in csv_reader:
        p=row[0]
        if p=='0':job_failed+=1
        else:job_passed+=1
    print("total number of jobs failed",job_failed)
    print("Total number of job passed",job_passed)
```

The column 1 of the Task_constraints file contains time
This column is mandatory field. 
The format of field is integer.

```python
# job id
d=dict()
c=0
lst=list()
with open('Task_Constraints.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=',')
    for row in csv_reader:
        p=row[1]
        if p not in d:d[p]=1
        else:d[p]+=1
    for key, val in list(d.items()):
        lst.append((val,key))
    lst.sort(reverse=True)
    for val1,key1 in lst:
        c=c+1
    print("total number of unique job_id :",c)
        
```

The column 2 of the Task_constraints file contains job id 
This column is mandatory field. 
The format of field is integer.

```python
# task_index
d=dict()
c=0
with open('Task_Constraints.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=',')
    for row in csv_reader:
        p=row[2]
        if p not in d:d[p]=1
        else:d[p]+=1
    for key,val in list(d.items()):
        c=c+1
    print("Total number of unique task index column :",c)
       
```

The column 3 of the Task_constraints file contains task_index
This column is mandatory field. 
The format of field is integer.

```python
# comparsion operator
import matplotlib.pylab as plt
d=dict()
lst=list()
with open('Task_Constraints.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=',')
    for row in csv_reader:
        p=row[3]
        if p not in d:d[p]=1
        else:d[p]+=1
    for key,val in list(d.items()):
        print(key,val)
    lst=sorted(d.items())
    x,y=zip(*lst)
    plt.bar(x,y)
    plt.show()
```

The column 4 of the Task_constraints file contains comparision operator
This column is mandatory field. 
The format of field is integer.

```python
# attribute value
import matplotlib.pylab as plt
d=dict()
keys=0
values=0
q=0
with open('Task_Constraints.csv') as csv_file:
    csv_reader = csv.reader(csv_file,delimiter=',')
    for row in csv_reader:
        p=row[5]
        if p!='':
            if p not in d:d[p]=1
            else:d[p]+=1
        else:q=q+1
    for key,val in list(d.items()):
        print(key,"     ",val)
    print("Total number of blank values ",q)
    keys=d.keys()
    values=d.values()
    plt.bar(keys,values)
```

This column is not mandatory field.
