## finding the time data for the machine_attribute column

```python
from matplotlib import pyplot as plt
from datetime import datetime
d=dict()
with gzip.open('/ichec/work/mucom001c/GoogleSvrLog/clusterdata-2011-1/machine_attributes/part-00000-of-00001.csv.gz','rb') as f:
    for line in f:
        line=line.decode()
        line=line.split(',')
        t=int(line[0])/1000
        date_time = datetime.fromtimestamp(t)
        d1=date_time.strftime("%d")
        if d1 not in d:d[d1]=1
        else:
            d[d1]+=1 
print(d)
lists = sorted(d.items()) # sorted by key, return a list of tuples
x, y = zip(*lists) # unpack a list of pairs into two tuples
plt.figure(figsize=(20,10))
plt.bar(x, y)
plt.xlabel('Time as date')
plt.ylabel('Count')
plt.title('Plot for Time as Date vs Count ')
plt.show()
```

## finding the attribute name for the machine_attribute column

```python
from matplotlib import pyplot as plt
dp=dict()
n=0
att_name=list()
with gzip.open('/ichec/work/mucom001c/GoogleSvrLog/clusterdata-2011-1/machine_attributes/part-00000-of-00001.csv.gz','rb') as f:
    for line in f:
        line=line.decode()
        line=line.split(',')
        p=line[2]
        tx=p
        if tx not in dp:dp[tx]=1
        else:
            dp[tx]+=1 
for k,v in dp.items():
    k=n
    att_name.append((k,v))
    n+=1
#print(c,"Are the total number of unique attrubute name in the dataset")
x, y = zip(*att_name) # unpack a list of pairs into two tuples
print(x,y)
plt.xlabel('Machine_attribute as number')
plt.ylabel('Count')
plt.title('Plot for machine_attribute vs Count ')
plt.plot(x, y)
plt.show()
```

## creating plot of machine_attribute with time column

```python
from matplotlib import pyplot as plt
from datetime import datetime
mt=dict()
n=0
tnn=dict()
ls=list()
with gzip.open('/ichec/work/mucom001c/GoogleSvrLog/clusterdata-2011-1/machine_attributes/part-00000-of-00001.csv.gz','rb') as f:
    for line in f:
        line=line.decode()
        line=line.split(',')
        t=int(line[0])/1000
        tn=line[2]
        date_time = datetime.fromtimestamp(t)
        d1=date_time.strftime("%d")
        ls.append((d1,tn))
x,y=zip(*ls)
fig, ax = plt.subplots(figsize=(20, 10))
ax.scatter(x, 
        y,
        marker = 'o',
        color = 'black')

# Set plot title and axes labels
ax.set(title = "attribute_name vs time",
       xlabel = "Time as date",
       ylabel = "attribute_name")

plt.show()
```
